```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")

model.train(
    data="/content/BIO-MEDICAL-WASTE--2/data.yaml",  # your correct dataset path
    epochs=50,
    imgsz=640,
    batch=16,
    name="biomedical_yolo_train"
)
```

```python
from google.colab import files
uploaded = files.upload()
```

```python

# âœ… Prediction without window popups (safe in Colab)
from matplotlib import pyplot as plt
import cv2

results = model.predict(source="/content/blood_cotton.webp", conf=0.2, show=False)

# Plot results inline
result_image = results[0].plot()
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
